```python
plt.scatter(df['age'],df['bought_insurance'],marker='v')

plt.xlabel('age')
plt.ylabel('buy insurance or not')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(df[['age']],df[['bought_insurance']], train_size=0.8)
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model_to_test=model.fit(x_train,y_train)
```

```python
model_to_test.predict(x_test)
```

```python
model_ready=model_to_test
```

```python
model_to_test.score(x_test,y_test)
```
