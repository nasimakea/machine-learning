```python
df=pd.read_csv("/content/py/ML/5_one_hot_encoding/Exercise/carprices.csv")
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['Car Model']=le.fit_transform(df['Car Model'])
```

```python
df
```

```python
x=df[['Car Model','Mileage','Age(yrs)']].values
x
```

```python
y=df['Sell Price($)'].values
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct=ColumnTransformer([('Car Model',OneHotEncoder(), [0])],remainder='passthrough')
```

```python
x=ct.fit_transform(x)
x
```

```python
x=x[:,1:]
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lg=LinearRegression()
```

```python
prdictor_model=lg.fit(x,y)
```

```python
prdictor_model.predict([[0, 3400, 2, 0]])
```
